# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

```python
Seaborn is a popular Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating informative and visually appealing statistical graphics. Here are five common plots that can be created using Seaborn along with their uses:

1)Scatter Plot:
Uses: Displays the relationship between two numerical variables. It helps identify patterns or trends, assess the strength of the correlation, and detect outliers.

2)Line Plot:
Uses: Shows the relationship between two continuous variables connected by a line. It is useful for visualizing trends and changes over time.

3)Bar Plot:
Uses: Compares categorical variables and displays their relative frequencies or aggregated values. It is suitable for visualizing comparisons, such as comparing sales by different categories.

4)Histogram:
Uses: Represents the distribution of a continuous variable by dividing it into bins and displaying the frequency or proportion of data points falling within each bin. It helps understand the shape, central tendency, and spread of the data.

5)Box Plot:
Uses: Summarizes the distribution of a continuous variable through quartiles, outliers, and other statistical measures. It provides insights into the range, skewness, and presence of outliers in the data.
```

# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
```

```python
fmri = sns.load_dataset('fmri')
```

```python
fmri
```

```python
sns.lineplot(x = fmri.timepoint , y = fmri.signal)
```

```python
sns.lineplot(x = fmri.timepoint , y = fmri.signal , data = fmri , hue ='event' , style='region')
```

# OR

```python
import seaborn as sns

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Create the line plot
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region")

# Display the plot
import matplotlib.pyplot as plt
plt.show()
```

# OR

```python
import seaborn as sns

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Create the first box plot: age vs. pclass
sns.boxplot(data=titanic_data, x="pclass", y="age")
plt.title("Age vs. Pclass")

# Display the first plot
import matplotlib.pyplot as plt
plt.show()

# Create the second box plot: fare vs. pclass
sns.boxplot(data=titanic_data, x="pclass", y="fare")
plt.title("Fare vs. Pclass")

# Display the second plot
plt.show()
```

# OR

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Plot histogram with hue
sns.histplot(data=diamonds_data, x='price', hue='cut')

# Display the plot
plt.title("Histogram of Diamond Prices by Cut")
plt.show()
```
